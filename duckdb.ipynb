```python
import duckdb
import pandas as pd

df = pd.read_csv("../2024q4/sub.txt", delimiter='\t')

# Connect to DuckDB (in-memory database by default)
con = duckdb.connect()

# Load the TSV file into a table
con.execute("CREATE TABLE my_table AS SELECT * FROM df")
```

```python
# Query the table to verify the data
result = con.execute("SELECT * FROM my_table limit 100").fetchall()
print(result)
```
